```python
movies_lst = [movies[movies.Genre == "Action"].BudgetMillions, movies[movies.Genre == "Drama"].BudgetMillions, movies[movies.Genre == "Thriller"].BudgetMillions, movies[movies.Genre == "Comedy"].BudgetMillions]

plt.hist(movies_lst, bins=15, stacked=True)
plt.show()
```

```python
# using a loop
list1 = []
my_labels = []
for gen in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre == gen].BudgetMillions)
    my_labels.append(gen)
h = plt.hist(list1, bins=30, stacked=True, rwidth=1,  label=my_labels)
plt.legend()
plt.show()
```

Kernel Density Estimation (KDE Plot)

```python
k1 = sns.kdeplot(x=movies.CriticRating, y=movies.AudienceRating, shade=True, thresh=0.05)
```

```python
sns.set_style('dark')
k1 = sns.kdeplot(x=movies.BudgetMillions, y=movies.AudienceRating, shade=True, thresh=0.05)
```

```python
k2 = sns.kdeplot(x=movies.BudgetMillions, y=movies.CriticRating, shade=True, thresh=0.05)
```
