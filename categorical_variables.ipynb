Coordinates and Diagonals

```python
# Controlling axes and adding a title
g = sns.FacetGrid(movies, row='Genre', col='Year', hue='Genre')
kws = dict(s=50, linewidth=0.5, edgecolor='black')
g = g.map(plt.scatter, 'CriticRating', 'AudienceRating', **kws)
g.set(xlim=(0,100), ylim=(0,100))
for ax in g.axes.flat:
    ax.plot((0,100), (0,100), c='grey', ls='--')
g.add_legend()
```

### Dashboards

```python
from matplotlib import pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
sns.set_style('darkgrid')
f, axes = plt.subplots(2, 2, figsize=(15,15))
k1 = sns.kdeplot(x=movies.BudgetMillions, y=movies.AudienceRating, shade=True, thresh=0.05, ax=axes[0,0])
k2 = sns.kdeplot(x=movies.BudgetMillions, y=movies.CriticRating, shade=True, thresh=0.05, ax=axes[0,1])
k4 = sns.boxplot(data=movies, x='Genre', y='CriticRating', palette='coolwarm', ax=axes[1,0])
k5 = sns.kdeplot(x=movies.CriticRating, y=movies.AudienceRating, shade=True, thresh=0.05, ax=axes[1,1])
plt.show()
```

Styling Dashboards

```python
sns.set_style('dark', {'axes.facecolor': 'black'})
f, axes = plt.subplots(2, 2, figsize=(15,15))

# Plot [0,0]
k1 = sns.kdeplot(x=movies.BudgetMillions, y=movies.AudienceRating, shade=True, thresh=0.05, ax=axes[0,0], shade_lowest=True, cmap='inferno')
k1 = sns.kdeplot(x=movies.BudgetMillions, y=movies.AudienceRating, thresh=0.05, ax=axes[0,0], cmap='cool')


# Plot [0,1]
k2 = sns.kdeplot(x=movies.BudgetMillions, y=movies.CriticRating, shade=True, thresh=0.05, ax=axes[0,1], cmap='inferno', shade_lowest=True)
k2b = sns.kdeplot(x=movies.BudgetMillions, y=movies.CriticRating, thresh=0.05, ax=axes[0,1], cmap='cool')

# Plot [1,0]
k3 = sns.boxenplot(data=movies, x='Genre', y='BudgetMillions', palette='YlOrRd', ax=axes[1,0])

# Plot [1,1]
k4 = sns.kdeplot(x=movies.CriticRating, y=movies.AudienceRating, shade=True, thresh=0.05, ax=axes[1,1], cmap='Blues_r', shade_lowest=True)
k4b = sns.kdeplot(x=movies.CriticRating, y=movies.AudienceRating, ax=axes[1,1], cmap='cool')
plt.show()
```
